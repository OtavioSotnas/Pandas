# Conhecendo a base de dados

## Importando dados

Principais funções para importar dados

- **read_csv()**: Essa função é usada para ler arquivos CSV (Comma Separated Values), que são arquivos de texto que contêm dados separados por vírgulas. É possível passar diversos parâmetros para personalizar a leitura do csv, como delimitador, cabeçalho, tipo de encoding, entre outros.

- **read_excel()**: Essa função é usada para ler arquivos do Excel (.xls ou .xlsx) e criar um DataFrame a partir dos dados.

- **read_json()**: Essa função é usada para ler arquivos JSON (JavaScript Object Notation), que são arquivos de texto que contêm dados em formato de objeto JavaScript.

- **read_html()**: Essa função é usada para ler tabelas HTML, que são estruturas de dados organizadas em formato de tabela em uma página da web.

- **read_sql()**: Essa função é usada para ler dados de um banco de dados relacional, como o MySQL, PostgreSQL e SQL Server. O Pandas é capaz de importar dados de diferentes formas, permitindo ajustar parâmetros como a consulta, o nome da tabela e o tipo de dados das colunas.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Armazenando o arquivo csv da nuvem na memória local
url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv"
```

```python
# Comando para ler um csv tendo o parametro de um csv
pd.read_csv(url)
```

```python
# Por padrão a função read_csv usa o argumento sep="," porem nosso arquivo está separado por ; e devemos mudar.
pd.read_csv(url, sep=";") 
```

```python
# Agora só precisamos declarar nossa base de dados em uma variável 
dados = pd.read_csv(url, sep=";") 
```

# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel ?

Resumo: 

- **df.groupby("coluna")**: realiza um agrupamento por alguma coluna específica, essa função funciona em conjunto com sum() ou mean() para realizar o agrupamento.

- **df.sort_values()**: ordena os valores em relação a um eixo.

```python
dados.head()
```

```python
# O comando groupby("Coluna") somente separa uma coluna do data frame, é necessário usar funções auxiliares como sum() ou mean() para sair os resultados.
dados.groupby("Tipo").mean(numeric_only=True).head() # A função mean() pega todas colunas inclusive as não numéricas, por isso o argumento "numeric_only=True"
```

```python
# Podemos aplicar o groupby e relacionar a apenas com uma ou mais colunas e ordenar usando o sort_values()
dados.groupby('Tipo')[["Valor"]].mean().sort_values("Valor")
```

```python
dados_preco_tipo = dados.groupby('Tipo')[["Valor"]].mean().sort_values("Valor")
```

```python
dados_preco_tipo.plot(kind="barh", figsize=(14,10), color="blue")
```

## Removendo os imóveis comerciais

Resumo: 

- **df["Coluna"].unique()**: retorna os valores unicos de uma coluna.

- **df.query("expressão")**: filtra linhas em um data frame utlizando uma expressão booleana em formato de string. Para usarmos uma variável na expressão necessário botar @ antes.

```python
# A função unique() retorna os valores unicos de uma coluna
dados["Tipo"].unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
# A função query() filtra linhas em um data frame utlizando uma expressão booleana em formato de string.
dados.query("@imoveis_comerciais in Tipo").head() # Para usarmos uma var na expressão necessário botar @ antes.
```

```python
# Assim separamos os dados em tipo comercial e não comercial para melhor visualização dos gráficos.
dados.query("@imoveis_comerciais not in Tipo").head()
```

```python
dados_imoveis_comerciais = dados.query("@imoveis_comerciais in Tipo")
dados_imoveis_residenciais = dados.query("@imoveis_comerciais not in Tipo")
```

```python
dados_imoveis_comerciais.groupby('Tipo')[["Valor"]].mean().sort_values("Valor").plot(kind="barh", figsize=(10,4), color="blue")
```

```python
dados_imoveis_residenciais.groupby('Tipo')[["Valor"]].mean().sort_values("Valor").plot(kind="barh", figsize=(10,4), color="blue")
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados ?

Resumo: 

- **df[Series].value_counts()**: retorna uma series somando os valores únicos, para transformar em % passamos o argumento "normalize=True".

- **df.to_frame()**: transforma uma series em um dataframe.

- **df.rename(columns={'nome_antigo': 'novo_nome'}, inplace=True)**: permite especificar um dicionário que associa o nome antigo da coluna ao novo nome que queremos atribuir.

```python
# value_counts() retorna uma series somando os valores únicos, para transformar em % passamos o argumento "normalize=True".
# to_frame() transforma uma series em um dataframe.
dados_percentual_tipo = dados_imoveis_residenciais['Tipo'].value_counts(normalize=True, ascending=True).to_frame()
dados_percentual_tipo
```

```python
dados_percentual_tipo.plot(kind="barh", figsize=(10,4), xlabel="Percentual", ylabel="Tipo", color="green")
```
